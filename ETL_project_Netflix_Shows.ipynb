# Import Libraries and Create File Path

```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
#pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
file = "datasets_1377_2482_Netflix Shows.csv"
df = pd.read_csv(file, encoding='latin-1')
```

```python
df
```

## Drop unnecessary columns 

```python
df = df.drop(['ratingDescription','release year', 'user rating size'], axis=1)
df
```

## Drop Null Values from "Title" and "Usering Rateing Score" columns 

```python
df.dropna(subset=['rating','user rating score'], inplace=True)
df
```

## Rename the Columns 

```python
df = df.rename(columns={'title':'title','rating':'age_rating','ratingLevel':'rating_level','user rating score':'user_rating'})
df
```

## Replace Other Null Values with Empty String

```python
df = df.replace(np.nan, '', regex=True)
df
```

## Make Sure the Data Type of Each Column is Correct

```python
df = df.astype({"title": str, "age_rating": str, "rating_level": str, "user_rating": int})
df
```

## Check for duplicates 'Titles' and double check the result

```python
df[df.duplicated('title')]
```

```python
df.loc[df['title'] == 'Grey\'s Anatomy']
```

```python
df.loc[df['title'] == 'Prison Break']
```

## Drop duplicates and reset index 

```python
df = df.drop_duplicates('title')
df.reset_index(drop=True, inplace=True)
df
```

## Create database connection 

```python
connection_string = "postgres:postgres@localhost:5432/ETL_Project"
engine = create_engine(f'postgresql://{connection_string}',pool_size=10, max_overflow=20)
```

```python
engine.table_names()
```

## Load DataFrames into SQL database

```python
df.to_sql(name='netflix_rating', con=engine, if_exists='append', index=False)
```
